```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "/content/BAC.csv"
df = pd.read_csv(file_path)
```

```python
sns.set_style("whitegrid")
```

```python
fig, axes = plt.subplots(2,2,  figsize=(15,12))
```

```python
plt.figure(figsize=(15, 6))
plt.plot(df['Date'], df['Close'], label="Closing Price", color="blue")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.title("Closing Prices Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
##Histogram
plt.figure(figsize=(8, 6))
plt.hist(df['Close'], bins=50, color='blue', alpha=0.7, edgecolor='black')
plt.title("Histogram of Closing Prices")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.show()
```

```python
##Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['High'], df['Low'], color='blue', alpha=0.5, edgecolors='black')

plt.title("Scatter Plot of High vs Low Prices", fontsize=14)
plt.xlabel("High Price", fontsize=12)
plt.ylabel("Low Price", fontsize=12)
plt.grid(True)

plt.show()
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

avg_volume_per_year = df.groupby('Year')['Volume'].mean()
```

```python
## BarPlot
 plt.figure(figsize=(10, 6))
plt.bar(avg_volume_per_year.index, avg_volume_per_year.values, color='orange', edgecolor='black')
plt.title("Average Trading Volume Per Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Average Volume", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  s

plt.show()
```

```python
# Compute skewness
skewness_value = skew(df['Close'].dropna())

# Plot histogram + KDE
plt.figure(figsize=(8, 6))
sns.histplot(df['Close'], bins=50, kde=True, color='blue', alpha=0.6)
plt.axvline(df['Close'].mean(), color='red', linestyle='dashed', label=f'Mean (Skewness = {skewness_value:.2f})')
plt.title("Histogram & KDE: Skewness Analysis", fontsize=14)
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.legend()
plt.grid()

plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 6))
plt.boxplot(df['Close'], vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'))
plt.title("Box Plot of Closing Prices", fontsize=14)
plt.ylabel("Closing Price", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

volume_per_year = df.groupby('Year')['Volume'].sum()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(volume_per_year, labels=volume_per_year.index, autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red', 'purple'],
        startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title("Percentage of Trading Volume by Year", fontsize=14)


plt.show()
```

```python
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

```python
# Plot violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(y=df['Close'], color="green", inner="quartile")

plt.title("Violin Plot of Closing Prices", fontsize=14)
plt.ylabel("Closing Price", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Sort values by date
df = df.sort_values(by='Date')

# Plot area chart
plt.figure(figsize=(10, 6))
plt.fill_between(df['Date'], df['Close'], color="skyblue", alpha=0.5)
plt.title("Area Chart of Closing Prices Over Time", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Closing Price", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python
from scipy.stats import skew
```

```python
# Compute skewness
skewness_value = skew(df['Close'].dropna())

# Plot histogram with KDE (Kernel Density Estimation)
plt.figure(figsize=(8, 6))
sns.histplot(df['Close'], bins=50, kde=True, color='blue', alpha=0.6)
plt.title(f"Skewness Chart of Closing Prices (Skewness = {skewness_value:.2f})", fontsize=14)
plt.xlabel("Closing Price", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.axvline(df['Close'].mean(), color='red', linestyle='dashed', label='Mean')
plt.legend()


plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['Close'], color="purple")

plt.title("Box Plot: Skewness Analysis", fontsize=14)
plt.ylabel("Closing Price")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
